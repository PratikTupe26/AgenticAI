```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsMan(TypedDict):
  four: int
  six:int
  run:int 
  balls:int
#   functions node data members
  SR:float
  BP:float
  BPB:float
summary:str
```

```python
# node defination 
def calculate_SR(state:BatsMan):
    sr = (state['run'] / state['balls'])*100
  
    return {'SR':sr}
     
def calculate_BP(state:BatsMan):
   bp = (state['four']*4 + state['six']*6)/100
   return {'BP':bp}

def calculate_BPB(state: BatsMan):
    total = state['four'] + state['six']
    bpb = state['balls'] / total if total else 0
    return {'BPB': bpb}

def summary(state:BatsMan):

   summary = f"""
strike rate of batsMan - {state['SR']}\n
boundary perentage - {state['BP']} \n
boundry per balls - {state['BPB']}

"""
  
   return {'summary':summary}
```

```python
# define graph

# define graph
graph = StateGraph(BatsMan)

# define nodes calls
graph.add_node('calculate_SR', calculate_SR)
graph.add_node('calculate_BP', calculate_BP)
graph.add_node('calculate_BPB', calculate_BPB)
graph.add_node('summary', summary)

# define edges 

graph.add_edge(START, 'calculate_SR')
graph.add_edge(START, 'calculate_BP')
graph.add_edge(START, 'calculate_BPB')

graph.add_edge('calculate_SR', 'summary')
graph.add_edge('calculate_BP', 'summary')
graph.add_edge('calculate_BPB', 'summary')
graph.add_edge('summary', END)


```

```python
workflow=graph.compile()
```

```python
initial_state = {
'four' : 4,
'six':6,
'run':200 ,
'balls': 90
}
result  = workflow.invoke(initial_state)
print(result)
```
